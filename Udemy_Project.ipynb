```python
# Q: plot of lectures per level per subject ?

def func_draw(z):
    plt.figure(figsize=(20,10))
    sns.barplot(x='level',y=z,data=df,hue='subject',errorbar=None)
    plt.title('plt for {} per level per subject'.format(z))
    plt.show
    #pie chart
    plt.figure(figsize=(5,5))
    df.groupby('level')[z].sum() .plot(kind='pie')
    plt.title('{} per level'.format(z))
    plt.show
```

```python
df['published_date'] = df['published_timestamp'].str.split('T').str.get(0)
```

```python
df['published_date']
```

```python
df['published_date'] =pd.to_datetime(df['published_date'],format='%Y-%m-%d',errors='coerce')

```

```python
df.dtypes
```

```python
df['published_date']
```

```python
df['year']=df['published_date'].dt.year.astype('Int64')
df['month']=df['published_date'].dt.month.astype('Int64')
df['day']=df['published_date'].dt.day.astype('Int64')
```
